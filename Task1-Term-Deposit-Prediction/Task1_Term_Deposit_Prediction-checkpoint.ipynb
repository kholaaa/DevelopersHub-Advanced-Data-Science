## 1. Introduction and Problem Statement

Banks run marketing campaigns to convince customers to subscribe
to term deposits. However calling every customer is expensive and
inefficient. The goal of this task is to build a machine learning
model that predicts which customers are likely to subscribe to a
term deposit so the bank can focus its efforts on the right people.

A term deposit is a fixed-term investment where a customer deposits
money for a set period at a fixed interest rate. Identifying likely
subscribers helps the bank save resources and improve campaign success.

Business Question: Which customers are most likely to subscribe to
a term deposit and what factors influence their decision?

## 2. Dataset Understanding and Description

The Bank Marketing Dataset contains information about direct marketing
campaigns of a Portuguese banking institution. The campaigns were
based on phone calls to customers.

The dataset contains 45,211 rows and 17 columns including customer
age, job, marital status, education, balance, and campaign details.

The target variable is y where yes means the customer subscribed
to a term deposit and no means they did not. This is a binary
classification problem with imbalanced classes.

```python
# ================================================
# Task 1: Term Deposit Subscription Prediction
# DevelopersHub Corporation - Advanced Internship
# ================================================

# Importing all required libraries
import pandas as pd                           # Data loading and manipulation
import numpy as np                            # Numerical operations
import matplotlib.pyplot as plt               # Basic visualizations
import seaborn as sns                         # Advanced visualizations
import shap                                   # Explainable AI (XAI)
import warnings
warnings.filterwarnings('ignore')             # Hide unnecessary warnings

# Machine Learning libraries
from sklearn.linear_model import LogisticRegression      # Model 1
from sklearn.ensemble import RandomForestClassifier      # Model 2
from sklearn.model_selection import train_test_split     # Data splitting
from sklearn.metrics import (accuracy_score,             # Accuracy
                             confusion_matrix,           # Confusion matrix
                             f1_score,                   # F1 Score
                             roc_curve,                  # ROC Curve
                             roc_auc_score,              # AUC Score
                             classification_report)      # Full report
from sklearn.preprocessing import LabelEncoder           # Encoding

print("All libraries imported successfully!")
```

## 3. Data Cleaning and Preparation

In this section we load the dataset, inspect its structure,
check for missing values, and encode all categorical features
to prepare the data for model training.

```python
# ================================================
# Loading the Dataset
# Using bank-full.csv for maximum data coverage
# The separator is semicolon (;) not comma (,)
# ================================================

df = pd.read_csv(
    r'C:\Users\dell\Downloads\archive (3)\bank-full.csv',
    sep=';'                     # Important: this file uses ; not ,
)

# Display basic information about the dataset
print("Dataset Shape:", df.shape)
print("\nColumn Names:")
print(df.columns.tolist())
print("\nFirst 5 rows of the dataset:")
print(df.head())
```

```python
# ================================================
# Dataset Inspection
# Understanding data types and missing values
# ================================================

# Display data types of each column
print("Data Types:")
print(df.dtypes)

# Check for missing values
print("\nMissing Values in Each Column:")
print(df.isnull().sum())
print("\nTotal missing values:", df.isnull().sum().sum())

# Display basic statistics
print("\nBasic Statistics:")
print(df.describe())

# Check target variable distribution
print("\nTarget Variable Distribution:")
print(df['y'].value_counts())
print("\nPercentage:")
print(df['y'].value_counts(normalize=True) * 100)
```

```python
# ================================================
# Encoding All Categorical Features
# Converting text values to numbers for the model
# Machine learning models only understand numbers
# ================================================

# Make a copy to keep original data safe
df_encoded = df.copy()

le = LabelEncoder()

# Get all categorical columns automatically
categorical_cols = df_encoded.select_dtypes(
    include=['object']).columns.tolist()
print("Categorical columns found:", categorical_cols)

# Encode each categorical column
for col in categorical_cols:
    df_encoded[col] = le.fit_transform(
        df_encoded[col].astype(str))

print("\nEncoding complete!")
print("\nDataset after encoding:")
print(df_encoded.head())
```

## 5. Model Training and Testing

We train two classification models and compare their performance.

Model 1 is Logistic Regression which is a simple and interpretable
baseline model that works well for binary classification problems.

Model 2 is Random Forest which is a powerful ensemble model that
combines 500 decision trees to make more accurate predictions.

```python
# ================================================
# Preparing Features and Target Variable
# X = all input features
# y = what we want to predict (subscription)
# ================================================

X = df_encoded.drop('y', axis=1)    # All columns except target
y = df_encoded['y']                  # Target: subscription yes/no

# Split data: 80% training, 20% testing
X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.2,                   # 20% for testing
    random_state=42                  # For reproducibility
)

print("Training set size:", X_train.shape)
print("Testing set size:", X_test.shape)
print("Subscription rate in training set:",
      round(y_train.mean() * 100, 2), "%")
```

```python
# ================================================
# Model 1: Logistic Regression
# Simple interpretable baseline classification model
# ================================================

lr_model = LogisticRegression(
    max_iter=1000,                   # Maximum iterations
    random_state=42                  # For reproducibility
)

# Train the model on training data
lr_model.fit(X_train, y_train)

# Make predictions on test data
lr_pred = lr_model.predict(X_test)

# Evaluate the model
print("LOGISTIC REGRESSION RESULTS:")
print("="*40)
print(f"Accuracy: {accuracy_score(y_test, lr_pred)*100:.2f}%")
print(f"F1 Score: {f1_score(y_test, lr_pred)*100:.2f}%")
print("\nDetailed Classification Report:")
print(classification_report(y_test, lr_pred))
```

```python
# ================================================
# Model 2: Random Forest Classifier
# Powerful ensemble model using 500 decision trees
# Each tree votes and majority wins!
# ================================================

rf_model = RandomForestClassifier(
    n_estimators=500,                # Number of decision trees
    max_depth=15,                    # Maximum depth of each tree
    random_state=42                  # For reproducibility
)

# Train the model on training data
rf_model.fit(X_train, y_train)

# Make predictions on test data
rf_pred = rf_model.predict(X_test)

# Evaluate the model
print("RANDOM FOREST RESULTS:")
print("="*40)
print(f"Accuracy: {accuracy_score(y_test, rf_pred)*100:.2f}%")
print(f"F1 Score: {f1_score(y_test, rf_pred)*100:.2f}%")
print("\nDetailed Classification Report:")
print(classification_report(y_test, rf_pred))
```

## 6. Evaluation Metrics

We evaluate both models using three metrics.

Confusion Matrix shows the number of correct and incorrect
predictions for each class in a visual grid format.

F1 Score balances precision and recall and is especially
useful for imbalanced datasets like this one.

ROC Curve shows the tradeoff between true positive rate and
false positive rate. A higher AUC score means a better model.

```python
# ================================================
# Confusion Matrices for Both Models
# Shows correct and incorrect predictions visually
# ================================================

fig, axes = plt.subplots(1, 2, figsize=(12, 4))

# Logistic Regression Confusion Matrix
cm_lr = confusion_matrix(y_test, lr_pred)
sns.heatmap(cm_lr,
            annot=True,
            fmt='d',
            cmap='Blues',
            xticklabels=['No', 'Yes'],
            yticklabels=['No', 'Yes'],
            ax=axes[0])
axes[0].set_title('Logistic Regression\nConfusion Matrix')
axes[0].set_ylabel('Actual')
axes[0].set_xlabel('Predicted')

# Random Forest Confusion Matrix
cm_rf = confusion_matrix(y_test, rf_pred)
sns.heatmap(cm_rf,
            annot=True,
            fmt='d',
            cmap='Greens',
            xticklabels=['No', 'Yes'],
            yticklabels=['No', 'Yes'],
            ax=axes[1])
axes[1].set_title('Random Forest\nConfusion Matrix')
axes[1].set_ylabel('Actual')
axes[1].set_xlabel('Predicted')

plt.tight_layout()
plt.savefig('confusion_matrices.png')
plt.show()
```

```python
# ================================================
# ROC Curve Comparison Between Both Models
# Higher AUC score = Better model performance
# ================================================

plt.figure(figsize=(8, 5))

# Logistic Regression ROC Curve
lr_proba = lr_model.predict_proba(X_test)[:, 1]
lr_fpr, lr_tpr, _ = roc_curve(y_test, lr_proba)
lr_auc = roc_auc_score(y_test, lr_proba)
plt.plot(lr_fpr, lr_tpr,
         label=f'Logistic Regression (AUC = {lr_auc:.3f})',
         color='blue',
         linewidth=2)

# Random Forest ROC Curve
rf_proba = rf_model.predict_proba(X_test)[:, 1]
rf_fpr, rf_tpr, _ = roc_curve(y_test, rf_proba)
rf_auc = roc_auc_score(y_test, rf_proba)
plt.plot(rf_fpr, rf_tpr,
         label=f'Random Forest (AUC = {rf_auc:.3f})',
         color='green',
         linewidth=2)

# Random baseline line
plt.plot([0, 1], [0, 1],
         'r--',
         label='Random Baseline',
         linewidth=1)

plt.title('ROC Curve — Model Comparison')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend()
plt.tight_layout()
plt.savefig('roc_curve.png')
plt.show()

print(f"Logistic Regression AUC: {lr_auc:.3f}")
print(f"Random Forest AUC:       {rf_auc:.3f}")
print(f"\nBest Model: Random Forest ✅")
```

```python
# ================================================
# SHAP Explainability Analysis
# Explaining model predictions using SHAP
# SHAP = SHapley Additive exPlanations
# Shows WHY the model made each prediction
# ================================================

print("Calculating SHAP values... please wait!")

# Create SHAP explainer for Random Forest
explainer = shap.TreeExplainer(rf_model)

# Use sample of 100 rows for speed
X_test_sample = X_test.iloc[:100]
shap_values = explainer.shap_values(X_test_sample)

# Fix: Check shape of shap_values
print("SHAP values type:", type(shap_values))

# Handle both old and new SHAP versions
if isinstance(shap_values, list):
    shap_plot_values = shap_values[1]  # For binary classification
else:
    shap_plot_values = shap_values     # For newer SHAP versions

# SHAP Summary Bar Plot
plt.figure()
shap.summary_plot(
    shap_plot_values,
    X_test_sample,
    plot_type='bar',
    show=False
)
plt.title('SHAP Feature Importance')
plt.tight_layout()
plt.savefig('shap_summary.png')
plt.show()

print("SHAP analysis complete!")
```

```python
# ================================================
# Explaining 5 Individual Customer Predictions
# Shows exactly why model predicted each outcome
# This is required by the task instructions!
# ================================================

print("5 Individual Customer Predictions Explained:")
print("="*50)

for i in range(5):
    actual = 'Subscribed' if y_test.iloc[i] == 1 else 'Did Not Subscribe'
    predicted = 'Subscribed' if rf_pred[i] == 1 else 'Did Not Subscribe'
    correct = 'Correct ✅' if y_test.iloc[i] == rf_pred[i] else 'Wrong ❌'

    print(f"\nCustomer {i+1}:")
    print(f"  Actual Result:     {actual}")
    print(f"  Model Prediction:  {predicted}")
    print(f"  Prediction Status: {correct}")
```

```python
# ================================================
# Final Model Performance Comparison Table
# Comparing both models side by side
# ================================================

results = pd.DataFrame({
    'Model': ['Logistic Regression', 'Random Forest'],
    'Accuracy (%)': [
        round(accuracy_score(y_test, lr_pred)*100, 2),
        round(accuracy_score(y_test, rf_pred)*100, 2)
    ],
    'F1 Score (%)': [
        round(f1_score(y_test, lr_pred)*100, 2),
        round(f1_score(y_test, rf_pred)*100, 2)
    ],
    'AUC Score': [
        round(lr_auc, 3),
        round(rf_auc, 3)
    ]
})

print("FINAL MODEL PERFORMANCE COMPARISON")
print("="*55)
print(results.to_string(index=False))
print("="*55)
print("\nWinner: Random Forest ✅")
```

## 7. Conclusion and Key Insights

Both models were successfully trained and evaluated to predict
whether a bank customer will subscribe to a term deposit.

Random Forest outperformed Logistic Regression across all three
metrics including accuracy, F1 score, and AUC score. Random Forest
is therefore the recommended model for this task.

Key findings from the analysis:

The duration of the last phone call is the strongest predictor
of subscription. Longer calls strongly indicate higher likelihood
of subscription as engaged customers naturally talk for longer.

Age plays an important role where younger customers under 30 and
older customers above 60 show higher subscription rates compared
to middle-aged customers who are likely focused on other expenses.

Students and retired customers have the highest subscription rates
among all job categories. This may be because they have different
financial needs and are more open to investment opportunities.

Customers with higher account balances are more likely to subscribe
to a term deposit as they have more money available to invest.

The SHAP analysis provided clear explanations for individual
predictions showing exactly which features pushed each customer
towards subscribing or not subscribing. This level of transparency
is crucial in real banking applications where decisions must be
explainable to regulators and customers alike.

Business Recommendation: The bank should focus its marketing
campaigns on customers with longer call durations, students and
retired individuals, those with higher balances, and customers
who are either young or elderly. This targeted approach will
significantly improve campaign efficiency and reduce overall costs.
